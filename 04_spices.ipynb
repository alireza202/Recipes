```python
import pickle
import sys, re
import urllib2
from scipy.sparse import lil_matrix
from sklearn.cluster import KMeans
```

```python
all_recipes = None
with open('recipes_simp_ing.pkl', 'rb') as infile:
    all_recipes = pickle.load(infile)
```

```python
list_ing = None
with open('simp_ing_dict.pkl', 'rb') as infile:
    list_ing = pickle.load(infile)
```

```python
# scraping the name of spices from a website
site = 'http://www.vegkitchen.com/tips/herbs-and-spices-common-culinary/'
regex1 = r'<p><strong>(.+)<\/strong>'

hdr = {'User-Agent': 'Mozilla/5.0'}
req = urllib2.Request(site,headers=hdr)
page = urllib2.urlopen(req).read()

spices = re.findall(regex1, page)
```

```python
# cleaning the list
spices = [x.lower() for x in spices if 'href' not in x]
spices.append('parsely')

spices[0] = 'anise'
spices[6] = 'cayenne'
spices[14] = 'curry'
spices[15] = 'dill'

spices.sort()
```

```python
# now we want to cluster recipes with similar spices
# we make a sparse matrix S with rows of recipes, and columns of spices
rdim = len(all_recipes)
cdim = len(spices)
S = lil_matrix((rdim, cdim))
```

```python
# filling in the sparse matrix with 1 where there is an spice
for i in range(rdim):
    for ing in all_recipes[i].simp_ing_name:
        if ing in spices:
            S[i, spices.index(ing)] = 1
```

```python
# perform clustering
n = 6
labeler = KMeans(n)
# convert lil to csr format
# note: Kmeans currently only works with CSR type sparse matrix
labeler.fit(S.tocsr()) 
```

```python
import pandas as pd
temp_list = list()

for (row, label) in enumerate(labeler.labels_):
    temp_list.append([all_recipes[row].name, label,
                      [x for x in all_recipes[row].simp_ing_name if x in spices]])

df = pd.DataFrame(temp_list, columns = ['name', 'cluster', 'ingredients'])
```

```python
# now we can finally see what spices go together!

from wordcloud import WordCloud
import matplotlib.pyplot as plt

plt.figure(figsize=(16,12))
for ii in range(n):
    text = ' '.join([x for l in list(df[df.cluster == ii].ingredients) if len(l) > 1 for x in l])
    wordcloud = WordCloud(font_path='/Users/alireza/Library/Fonts/Verdana.ttf').generate(text)

    # Open a plot of the generated image.
    
    plt.subplot(3, 2, ii+1)
    plt.imshow(wordcloud)
    plt.axis("off")

plt.show()
```
